# Financial Performance Comparison:IBM Competitors Analysis(2015-2024)

## Problem Definition
**Target User**:Investment analyst or retail investor.
**Analytical Question**:Which company has generated higher net income over the past 10 years?
**Business Value**:It helps investors decide which technology stock has shown stronger profitability.

## Data Source
-Source:WRDS Compustat(funda table)
-Data Accessed:April 2026

```python
# Import libraries
import wrds
import pandas as pd
import matplotlib.pyplot as plt

#Connect to WRDS
db = wrds.Connection()
print("Connected to WRDS successfully!")
```

## Step1:Query Data from WRDS
The SQL query below retrieves net incpme data for two IBM related companies from the compustat database for fiscal years 2015 to 2024.

```python
#SQL query
query = """
    SELECT
        gvkey,
        conm,
        fyear,
        ni,
        datadate
    FROM comp.funda
    WHERE gvkey IN('001000', '006066')
      AND fyear BETWEEN 2015 AND 2024
      AND indfmt = 'INDL'
      AND datafmt = 'STD'
      AND consol = 'C'
      AND popsrc = 'D'
    ORDER BY gvkey,fyear
"""

print("Running query on WRDS Compustat")
```

```python
#Execute query
financial_data = db.raw_sql(query)

print(f"Data retrieved: {len(financial_data)} rows")
print("\nFirst 5 rows:")
print(financial_data.head())
```

```python
#Close WRDS connection
db.close()
print("WRDS connection closed")
```

## Step 2: Prepare Comparison Table
The data is pivoted so that years become rows and companies become columns.Net income values are converted to billions of US dollars for easier reading.

```python
#Create comparison table
comparison = financial_data.pivot(index='fyear',columns='conm',values='ni')
comparison.columns = comparison.columns.str.strip()

#Convert to billions (divide by 1000)
comparison_billions = comparison / 1000
comparison_clean = comparison_billions.dropna(how='all')

print("Net Income Comparison (Billions USD):")
print(comparison_clean.round(2))
```

## Step 3: Bar Chart Comparison
The bar chart below shows IBM's annual net income performance for each fiscal year.

**How to read this chart:**
- Blue bars = IBM(INTL BUSINESS MACHINES CORP)
- Taller bar = higher annual net income
- Numbers on top show the exact value in billios USD

**What to look for:**
- Which company has taller bars in most years?
- Is the gap between them growing or shrinking?

```python
# Bar chart
fig,ax = plt.subplots(figsize=(12, 7))

comparison_clean.plot(kind='bar',ax=ax,width=0.7,color=['#1f77b4'])

ax.set_title('IBM Annual Net Income Comparison (2015-2024)\nSource: WRDS Compustat',
             fontsize=16,fontweight='bold')
ax.set_xlabel('Fiscal Year', fontsize=12)
ax.set_ylabel('Net Income (Billions USD)',fontsize=12)
ax.legend(fontsize=12)
ax.axhline(y=0, color='black', linewidth=0.8)
ax.grid(axis='y', alpha=0.3)

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='$%.2fB', fontsize=9, padding=3)

plt.tight_layout()
plt.show()
```

## Step 4:Growth Trend Analysis

The line chart below shows how IBM's net income has changed over the full 10-year analysis period.

**How to read this chart:**
- Blue line = IBM (INTL BUSINESS MACHINES CORP)
- Upward slope = annual profits are growing
- Downward slope = annual profits are declining
- Steeper slope = faster profit fluctuation

**What to look for:**
- Overall long-term profit trend across the decade
- Years with obvious profit rises or drops
- Stability and volatility of annual earnings

```python
# Line chart
fig, ax = plt.subplots(figsize=(12, 7))

comparison_clean.plot(kind='line',marker='o',ax=ax, linewidth=2.5, markersize=8)

ax.set_title('IBM Net Income Growth Trend (2015-2024)\nSource: WRDS Compustat',
             fontsize=16,fontweight='bold')
ax.set_xlabel('Fiscal Year',fontsize=12)
ax.set_ylabel('Net Income (Billions USD)', fontsize=12)
ax.legend(fontsize=12)
ax.grid(True,alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 5: Summary Statistics
The table below provies key statistics: average, maximum, minimum, and standard deviation of IBM's net income over the 10-yaer analysis period.

```python
# Summary statistics
summary = comparison_clean.describe()

print("SUMMARY STATISTICS (Billions USD)")
print("="*50)
print(summary.round(2))

company_name = comparison_clean.columns[0]

# key findings
avg_ibm = comparison_clean[company_name].mean()
total_ibm = comparison_clean[company_name].sum()
max_ibm = comparison_clean[company_name].max()
min_ibm = comparison_clean[company_name].min()

print("\nKEY FINDINGS:")
print("="*50)
print(f"Average Annual Net Income - {company_name}: ${avg_ibm:.2f} billion")
print(f"Total 10-year Total Net Income: ${total_ibm:.2f} billion")
print(f"Highest Annual Net Income: ${max_ibm:.2f} billion (2015)")
print(f"Lowest Annual Net Income: ${min_ibm:.2f} billion (2022)")

print(f"\nConclusion: Throughout 2015 to 2024, IBM maintained stable positive net income in all fiscal years.The company reached its peak profit in 2015, with moderate earnings fluctuations in the following decade, showing typical profitability characteristics of a mature global technology enterprise.")
```
